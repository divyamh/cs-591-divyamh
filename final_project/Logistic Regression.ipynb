```python
# We start project here
# First we will retrieve the movies from movie database.
# This will retreive all the movies from 1m Database and put it into a dataframe
# We will not be able to obtain the imdb ids for all the movies in the database.
# We will remove items for which IMDB,TMDB ID does not exist.
movielist = []
with open("./ml-1m/movies.dat") as moviefile:
    for line in moviefile:
        movie = line.split('::')
        movielist.append(movie)

movie1mdf = pd.DataFrame(movielist,columns = ['id','name','Genre'])
movie1mdf
```

```python
with open('./ml-20m/movies.csv') as movief:
    reader = csv.reader(movief)
    largemovielist = list(reader)
      
largemovielist.pop(0)
largemoviedf = pd.DataFrame(largemovielist, columns = ['id','name','genre'])
largemoviedf
```

```python
# The below code gets the year of release for largemoviedf from it's name
def getYearOfRelease(moviedf):
    yearnotavailable = 0
    yearofreleaselist = []
    for elem in moviedf['name']:
        namelist = elem.split('(')
        if(len(namelist) == 3):
            yearofrelease = namelist[2].split(')')[0]
        if(len(namelist) == 2):
            yearofrelease = namelist[1].split(')')[0]
        if(len(namelist) == 1):
            yearofrelease = 0
            yearnotavailable += 1
        yearofreleaselist.append(yearofrelease)
    print yearnotavailable
    return yearofreleaselist
```

```python
largemoviedf['release_year'] = getYearOfRelease(largemoviedf)
movie1mdf['release_year'] = getYearOfRelease(movie1mdf)
```

```python
largemoviedf['name'] = largemoviedf['name'].apply(lambda x:x.split('(',1)[0])
movie1mdf['name'] = movie1mdf['name'].apply(lambda x:x.split('(',1)[0])
```

```python
# Now we have two list of movies and the second list of movie has the id for which we have imdb id data
# We will take large movie list and create a dictionary with keys and values, where key is the movie name and value is
#id.
#From the second list will be obtain the imdb id and TMDB id
type(largemoviedf['name'])
movie1mdf['name_year'] = movie1mdf['name']+ movie1mdf['release_year'].map(str)
largemoviedf['name_year'] = largemoviedf['name']+largemoviedf['release_year'].map(str)
namelist = largemoviedf['name_year'].tolist()
idlist = largemoviedf['id'].tolist()
moviedictionary = dict(zip(namelist, idlist))
```

```python
# This provides the list of the movies for which we cannot get IMDB ID and TMDB ID.
movieid2 = []
count = 0
for elem in movie1mdf['name_year']:
    try:
        movieid2.append(moviedictionary[elem])
    except:
        print "%s \n", elem
        count+=1
print "The count of movies not found is", count
# We will remove the list of 183 movies for which we cannot find the imdb id.
```
