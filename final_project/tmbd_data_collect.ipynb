```python
import csv
import pandas as pd

with open('completedata.csv') as movief:
    reader = csv.reader(movief)
    largemovielist = list(reader)
      
largemovielist.pop(0)
largemoviedf = pd.DataFrame(largemovielist, columns = ['num','id','name','year','name_year','tmdbid','imdbid','budget','revenue','collection','popularity','cast'])
largemoviedf
```
