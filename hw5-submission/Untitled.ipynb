```python
import pandas as pd
import json
import numpy as np
from collections import defaultdict

def readJson(f):
    for l in open(f):
        yield eval(l)

Train = []
for l in readJson('train.json'):
    user,item,rating = l['reviewerID'],l['asin'],float(l['overall'])
    data =[user,item,rating]
    Train.append(data)

df = pd.DataFrame(Train, columns = ['ReviewerID','ItemID','Score'])
```
